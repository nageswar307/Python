```python
import requests

response = requests.get("https://www.gutenberg.org/cache/epub/100/pg100.txt")

shakespears_text = response.text
```

```python
len(shakespears_text)
```

```python
shakespears_text = shakespears_text[:10000]
```

```python
# pre processing

shakespears_text = shakespears_text.lower()
shakespears_text[:100]
```

```python
shakespears_text = shakespears_text.replace("\n"," ").replace("\r"," ").replace("***"," ").replace(".","").replace(","," ").replace("?"," ").replace("!","").replace("\\u","")
shakespears_text
```

```python
words = shakespears_text.split(" ")
```

```python
words
```

```python
words = [word for word in words if word != ""]
words
```

```python
stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
```

```python
len(stop_words)
```

```python
words = [word for word in words if word not in stop_words]
words
```

```python
len(words)
```

```python
len(set(words))
```

```python
word_freq = {}
```

```python
for word in words:
    if word not in word_freq:
        word_freq[word] = 1
    else:
        word_freq[word] += 1
```

```python
word_freq
```

```python
len(word_freq)
```

```python
sorted_words = sorted(word_freq.items(),key=lambda x:x[1],reverse=True)
```

```python
sorted_words[8]
```

```python
#bigram

bigram_freq = {}


for i in range(len(words)-1):
    bigram = (words[i], words[i+1])
    bigram_freq[bigram] = bigram_freq.get(bigram,0) + 1



sorted_bigrams = sorted(bigram_freq.items(),key=lambda x:x[1],reverse=True)
len(sorted_bigrams)
```

```python
input = "king"


prediction = []

for bigram in sorted_bigrams:
    if bigram[0][0] == input:
        prediction.append((bigram[0][1],bigram[1]))
```

```python
prediction
```

```python
next_word_dict = {}

for i in range(len(words)-1):
    current_word = words[i]
    next_word = words[i+1]

    if current_word not in next_word_dict:
        next_word_dict[current_word] = {}
        next_word_dict[current_word][next_word] = next_word_dict[current_word].get(next_word,0)+1
```

```python
next_word_dict
```

```python
predict_for = "king"
```

```python
if predict_for in next_word_dict:
    predictions = next_word_dict[predict_for]
    print(f"{predictions=}")
    prediction = max(predictions,key=predictions.get)
    print(f"{prediction=}")
```
